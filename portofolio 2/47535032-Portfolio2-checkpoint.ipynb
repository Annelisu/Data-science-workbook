```python
from sklearn.preprocessing import OrdinalEncoder 

# ensure to convert into an integer
Numerical_data = OrdinalEncoder(dtype = int)
read_data["num_fuel"] = Numerical_data.fit_transform(read_data[["fuel"]])
read_data["num_seller_type"] = Numerical_data.fit_transform(read_data[["seller_type"]])
read_data["num_transmission"] = Numerical_data.fit_transform(read_data[["transmission"]])
read_data["num_owner"] = Numerical_data.fit_transform(read_data[["owner"]])

read_data[[ "num_fuel", "num_seller_type", "num_transmission", "num_owner", "fuel", "owner", "seller_type", "transmission"]]

```

```python
# Calculate the correlation between year/km_driven/seller_type/fuel/owner and selling_price
year_corr = read_data['year'].corr(read_data['selling_price'])
km_driven_corr = read_data['km_driven'].corr(read_data['selling_price'])
seller_type_corr = read_data['num_seller_type'].corr(read_data['selling_price'])
fuel_corr = read_data['num_fuel'].corr(read_data['selling_price'])
owner_corr = read_data['num_owner'].corr(read_data['selling_price'])

```

### Q4: Train Linear Regression Models with Feature Selection under Cases 1 & 2
* When training a machine learning model for prediction, we may need to select the most important/correlated input features for more accurate results.
* To investigate whether feature selection affects the model performance, please select two most correlated features and two least correlated features regarding selling_price, respectively.
* Train four linear regression models by following the conditions:
    - (model-a) using the training/testing data in case 1 with two most correlated input features
    - (model-b) using the training/testing data in case 1 with two least correlated input features
    - (model-c) using the training/testing data in case 2 with two most correlated input features
    - (model-d) using the training/testing data in case 2 with two least correlated input features
* By doing this, we can verify the impacts of the size of traing/testing data on the model performance via comparing model-a and model-c (or model-b and model-d); meanwhile the impacts of feature selection can be validated via comparing model-a and model-b (or model-c and model-d).    

```python
# Train and test the 4 linear regression models
# Model-a
reg_model1 = linear_model.LinearRegression()
x_train_model1 = train_1[['year', 'num_fuel']]
y_train_model1 = train_1['selling_price']

x_test_model1 = test_1[['year', 'num_fuel']]
y_test_model1 = test_1['selling_price']

reg_model1.fit(x_train_model1, y_train_model1)
```

```python
# Model-b
reg_model2 = linear_model.LinearRegression()
x_train_model2 = train_1[['km_driven', 'num_owner']]
y_train_model2 = train_1['selling_price']

x_test_model2 = test_1[['km_driven', 'num_owner']]
y_test_model2 = test_1['selling_price']


reg_model2.fit(x_train_model2, y_train_model2)
```

```python
# Model-c
reg_model3 = linear_model.LinearRegression()
x_train_model3 = train_2[['year', 'num_fuel']]
y_train_model3 = train_2['selling_price']

x_test_model3 = test_2[['year', 'num_fuel']]
y_test_model3 = test_2['selling_price']

reg_model3.fit(x_train_model3, y_train_model3)
```

```python
# Model-d
reg_model4 = linear_model.LinearRegression()
x_train_model4 = train_2[['km_driven', 'num_owner']]
y_train_model4 = train_2['selling_price']

x_test_model4 = test_2[['km_driven', 'num_owner']]
y_test_model4 = test_2['selling_price']

reg_model4.fit(x_train_model4, y_train_model4)
```

### Q5: Evaluate Models
* Evaluate the performance of the four models with two metrics, including MSE and Root MSE
* Print the results of the four models regarding the two metrics

```python
# retrive the results of the four models using MSE and root MSE

# model A performance - case 1, most correlated feature
predict_a = reg_model1.predict(x_test_model1)
mse_a =((np.array(y_test_model1)-predict_a)**2).sum()/len(y_test_model1)


print("Model A MSE:", mse_a)
print("Model A root MSE:", np.sqrt(mse_a))

```

```python
# model B performance - case 1, least correlated feature
predict_b = reg_model2.predict(x_test_model2)
mse_b =((np.array(y_test_model2)-predict_b)**2).sum()/len(y_test_model2)

print("Model B MSE:", mse_b)
print("Model B root MSE:", np.sqrt(mse_b))
```

```python
# model C performance - case 2, most correlated feature
predict_c = reg_model3.predict(x_test_model3)
mse_c =((np.array(y_test_model3)-predict_c)**2).sum()/len(y_test_model3)

print("Model C MSE:", mse_c)
print("Model C root MSE:", np.sqrt(mse_c))
```

```python
# model D performance - case 2, least correlated feature
predict_d = reg_model4.predict(x_test_model4)
mse_d =((np.array(y_test_model4)-predict_d)**2).sum()/len(y_test_model4)

print("Model C MSE:", mse_d)
print("Model C root MSE:", np.sqrt(mse_d))
```

### Q6: Visualize, Compare and Analyze the Results
* Visulize the results, and perform ___insightful analysis___ on the obtained results. For better visualization, you may need to carefully set the scale for the y-axis.
* Normally, the model trained with most correlated features and more training data will get better results. Do you obtain the similar observations? If not, please ___explain the possible reasons___.

```python
# Creating dataframe from the MSE and Root MSE values
result = {
    'Models' : ['A', 'B', 'C', 'D'],
    'MSE' : [mse_a, mse_b, mse_c, mse_d],
    'Root MSE' : [np.sqrt(mse_a),np.sqrt(mse_b),np.sqrt(mse_c),np.sqrt(mse_d)],
}

df = pd.DataFrame(result)

# Plotting for MSE and Root MSE
sns.set(style="whitegrid")
plt.figure(figsize=(14, 5))

# Plotting for MSE
plt.subplot(1, 2, 1)
sns.barplot(x='Models', y='MSE', data=df)
plt.title('Comparison on MSE', size=18)
plt.xlabel('Models', size=15)
plt.ylabel('MSE', size=15)

# Plotting for Root MSE
plt.subplot(1, 2, 2)
sns.barplot(x='Models', y='Root MSE', data=df, color='lightcoral')
plt.title('Comparison on Root MSE', size=18)
plt.xlabel('Models', size=15)
plt.ylabel('Root MSE', size=15)

plt.tight_layout()
plt.show()
```

### Mean squared error (MSE) Models :
Model B has the highest mean squared error compared to the other models, which means its predictions are the furthest from the actual values. In contrast, Model C has the lowest MSE, indicating that its predictions are the most accurate among the models.

### Root mean squared error (Root MSE) Models :
Similarly to the MSE, Model C also has the lowest root mean squared error. The Root MSE represents the average error in the original units. However, all models have Root MSE values above 100,000, meaning that, on average, the predictions are quite far from the actual car prices. This suggests that the overall models' predictions are not very accurate, and further refinement is necessary to improve prediction accuracy.

Even though the Root MSE values are high, Model C was trained using the most correlated features and the best training data compared to the other models. Therefore, it demonstrates better and more accurate results based on the observations.

# Ethical Analysis
The infographic displays two tables of the 2008 Summer Olympic Medals, sorted either by total medals or by gold medals. Although both tables accurately represent the medals each country received, the choice of sorting impacts the viewer's perception. In Table 1, the sorting is based on total medals, placing the United States on top. However, in Table 2, where the medals are sorted by gold medals, China is placed on top.

The ethical concern lies in the potential for each table to promote different viewpoints and potentially downplay or ignore other relevant information. For instance, if only one table is shown, it could lead to a biased interpretation of which country performed better. This selective presentation can deceive the audience and downplay the representation of other countries. Ethical data visualization is crucial, especially in emphasizing fairness and objectivity to prevent the presentation of partial data and biases. Providing transparency and unbiased data sources will allow the audience to form an informed opinion. Therefore, showing both tables together and discussing how the medals are sorted will help mitigate the risk of misleading the audience.
